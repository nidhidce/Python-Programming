### Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

```python
#using the math library
# first take a coardinate from users
import math
x1=int(input("Enter the x coordinate of 1st point : "))
y1=int(input("Enter the y coordinate of 1st point : "))
x2=int(input("Enter the x coordinate of 2st point : "))
y2=int(input("Enter the y coordinate of 2st point : "))
#apply dis formula
dis= math.sqrt((x1-x2)**2+(y1-y2)**2)
# print result
print("distance between point is ", dis)



```

### Q7:- Write a program to find the sum of squares of first n natural numbers where n will be provided by the user.

```python
# with loop

num= int(input("Enter a nth number for finding sum of squares of first n natural number"))
sum=0
if num > 0:
  for i in range(1, num+1):
    sum=sum+(i*i)

  print("sum of square of first n natural numbers : ", sum)

else:
  print("your input is negative or 0")
```

```python
#direct formula (without loop)
n = int(input("enter a nth number for finding sum of squares of first n natural number "))
sum = (n*(n+1)*(2*n+1))//6
print(sum)
```

### Q8:- Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

```python
# taking 1st , 2nd and nth number from users
# finding difference (2nd-1st)
# apply formula for finding nth term.

a1=int(input("enter 1st term of an Arithmetic Series : "))
a2=int(input("enter 2nd term of an Arithmentic series : "))
n= int(input("enter the Nth number : "))
d=a2-a1
an = a1 + (n-1)*d
print("Nth term of the series : ", an)
```

66###Q10:- Given the height, width and breadth of a milk tank, you have to find out how many glasses of milk can be obtained? Assume all the inputs are provided by the user.

Input:
Dimensions of the milk tank
H = 20cm, L = 20cm, B = 20cm

Dimensions of the glass
h = 3cm, r = 1cm

```python
# taking all input form users
import math
H=float(input("enter the height of milk tank: "))
L=float(input("Enter the length of milk tank: "))
B=float(input("Enter the Breath of milk tank: "))

h=float(input("enter the height of glass: "))
r=float(input("enter the radius of glass: "))


volumn_milk_tank = H*L*B
volumn_glass=3.14*r*r*h

no_of_glass=(volumn_milk_tank)/(volumn_glass)
print("Total number of glass of milk = ",math.floor(no_of_glass)) ## taking floor value
```
